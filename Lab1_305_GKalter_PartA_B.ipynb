A "Moving Average" is a way of smoothing out a series of numbers to make trends easier to see.  Have a look at:  https://www.worldometers.info/coronavirus/usa/maine/ and look at the bar graphs toward the bottom of the page.  Try turning on the 3-day and 7-day moving averages to see the smoother versions of the data.  Now we're going to do that with our data.  Here's an example of how to do this, examine this code closely:

```python
import numpy as np
data = np.asarray([1,2,3,4,5,4,3,2,1])
window_length = 3
data
```

```python
extended_data = np.insert(data, 0, 0)
extended_data
```

```python
cum_sums = np.cumsum(extended_data)
cum_sums
```

```python
diffs = cum_sums[window_length:] - cum_sums[:-window_length]
diffs
```

```python
moving_averages = diffs / window_length
moving_averages
```

```python
import matplotlib.pyplot as plt
plt.rc('figure', figsize=(10, 6))
plt.title("Window=4") 
plt.bar(np.arange(len(data)), data, width = 0.1, align = 'center') 
# add reasonable values at both ends to make things line up correctly
moving_averages = np.insert(moving_averages, 0, 1)
moving_averages = np.append(moving_averages, 1)
plt.plot(moving_averages)
```
